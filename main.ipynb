### Functions that create players

```python
def generate_players(amount):
    player_list={}
    for i in range(0,amount):
        player_list[i]=Player()
    return player_list

def generate_cards():
    bank_card_list={}
    for i in range(0,40):
        if i%10==5:
            bank_card_list[i]=Card(location=i, rent=[25,50,100,200], cost=200)
            bank_card_list[i].type['isRailway']=True
        
        elif i in card_dictionary:
            bank_card_list[i]=Card(location=i, rent=card_dictionary[i]['rent'], cost=card_dictionary[i]['cost'],color=card_dictionary[i]['color'])
        
        else:
            bank_card_list[i]=Card(location=i)
            
    return bank_card_list
    
```

### Defining the Environment (Game)

```python
class Game:
    def __init__ (self, players=4):
        self.players=generate_players(players)
        self.cards=generate_cards()
        self.registry={}
        self.bank={0:{1,2,3,4}}
        self.turn=0
        #Ideally this should be one dictionary like we did for bank, I will know for the next game. We are way too far down, we just need results
        self.registryHistory={}
        self.locations={0:0,1:0,2:0,3:0}
        self.locationsHistory={}
        
    
    
    def play_turn(self):
        self.registryHistory[self.turn]=dict(self.registry)
        self.turn+=1
        self.bank[self.turn]={}
        self.locationsHistory[self.turn]=dict(self.locations)
        
        for player in self.players:
            if self.players[player].isBankrupt:
                continue
            #Roll
            playerRoll=[random.randrange(1,7),random.randrange(1,7)]
            self.players[player].location+=sum(playerRoll)
            #
            if self.players[player].location>=40:
                self.players[player].round+=1
                self.players[player].location=self.players[player].location%40
                self.players[player].cash+=5
            location=self.players[player].location
            
            #Only worked with try close, something with the comparator. Go to investigate more to fix with only if statements
            try:
                if self.players[player].cash>self.cards[location].cost:
                    self.players[player].cash-=self.cards[location].cost
                    self.players[player].buy_card(self.cards[location])
                    del self.cards[location]
                    self.registry[location]=player
                
            except:
                self.pay_rent(player, self.registry[location])
            
            self.bank[self.turn][player]=self.players[player].cash
            
            if self.players[player].cash<0:
                self.cards.update(self.players[player].cards)
                for i in self.players[player].cards:
                    del self.registry[i]
                
                self.players[player].isBankrupt=True
                print(f'Player {player} is bankrupt on turn {self.turn}!' )
            
            self.locations[player]=location
        
    def pay_rent(self, tenant1, host1):
        tenant=self.players[tenant1]
        host=self.players[host1]
        rentlevel=0
        if tenant.location%10==4:
            for i in host.cards:
                if host.cards[i].type['isRailway']:
                    rentlevel+=1
                    print(host1)
                    
        rent=host.cards[tenant.location].rent[rentlevel]
        self.players[tenant1].cash-=rent
        self.players[host1].cash+=rent
        #print(f'{tenant1} pay {host}')
    
            
```
